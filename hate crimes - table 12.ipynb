```python
#### IMPORTANT
# Because the number of agencies reporting in a given year vary, but the "population covered"
# column in this table is the whole state population, i instead want to use the total number 
# of populations per agency based on Table 13. However, that's skewed as well, since it does
# not include towns that participated but reported no hate crimes, which was about half of 
# all agencies. Both methodologies are imperfect.

# Calculations from my other sheet: 
#2014 2210928.0
#2011 1785489.0
#2010 1638134.0
#2013 1970207.0
#2012 1974862.0

effective_pops = {'2010': 1638134.0,
 '2011': 1785489.0,
 '2012': 1974862.0,
 '2013': 1970207.0,
 '2014': 2210928.0,
 '2015_unadjusted': 2346897.0,
 '2015': 1822339.0}
```

```python
def effective_per_capita(df):
    tmp_df = df.copy();
    year = tmp_df["year"].max()
    tmp_df["adj_per_cap"] = tmp_df["Total\nnumber of\nincidents\nreported"] * 1000 * 1000 / effective_pops[year]

    tmp_df["adj_per_cap_rank"] = tmp_df["adj_per_cap"].rank(ascending=False)
    return tmp_df.sort_values(by="adj_per_cap",ascending=False)

def per_capita (df):
    tmp_df = df.copy();
    tmpdf = effective_per_capita(tmp_df)
    tmp_df["per_cap"] = tmp_df["Total\nnumber of\nincidents\nreported"] * 1000 * 1000 / tmp_df["Population\ncovered"]

    tmp_df["per_cap_rank"] = tmp_df["per_cap"].rank(ascending=False)
    
    return tmp_df.sort_values(by="per_cap",ascending=False)

def state_only(df, state):
    return df[df["Participating state"]==state]


def ct_only(df):
    return state_only(df, "Connecticut")

def agencies_participating_only(df):
    return df["Number of\nparticipating\nagencies"]

def agencies_reporting_only(df):
    return df["Agencies\nsubmitting\nincident\nreports"]

def percap_only(df):
    return df["per_cap"]

def rank_only(df):
    return df["per_cap_rank"]

def state_rank(df,state):
    return rank_only(state_only(df,state))

def state_percap(df,state):
    return percap_only(state_only(df,state))

def ct_rank(df):
    return int(rank_only(state_only(per_capita(df),"Connecticut")))

def ct_percap(df):
    return int(percap_only(state_only(per_capita(df),"Connecticut")))

def ct_participating(df):
    return int(agencies_participating_only(state_only(per_capita(df),"Connecticut")))

def ct_reporting(df):
    return int(agencies_reporting_only(state_only(per_capita(df),"Connecticut")))
```

### CT conclusions from table 12

Based on the above analysis: 

The number and rate of reported hate crime incidents in CT were at a low in 2015 since data became available in 2010 and the state's rank improved over that time. However, the number of police agencies participating dropped from 102 agencies to 95 in Connecticut. The number of agencies reporting incidents did not decline over that period. The number of agencies reporting hate crime incidents ranged from 42 to 54 agencies in 2014. In 2015, it was 44.


### But...

As described above, it turns out the per_capita information from the feds is based on the entire state population, even though about 100 departments participate in this reporting program and only about 40-50 towns report at least hate crime. And because the number of agencies reporting changes from year to year, we don't have a constant baseline.

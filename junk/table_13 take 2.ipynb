```python
def tot_by_type(df, year):
    tmpdf = df.copy()
    tmpdf["year"] = year
    return df[df["Unnamed: 0"] == "Total"]

def get_tots():
    frames = []
    for k in table_13:
        df = table_13[k].copy()
        df["year"] = k
        frames.append(tot_by_type(df,k))
        
        
    ret = pd.concat(frames).set_index("year").transpose().drop("Unnamed: 0")
    return ret

get_tots()

def year_tot(year):
    df = get_tots()
    return df[year].sum()

for year in range(2010,2016):
    print year, year_tot(str(year))
```

### Trying to per-capita-ize the cities

```python
# Print the sum population of all the reporting cities 
def pop(year):
    return cities(year)["pop"].sum()

years = [2010,2011,2012,2013,2014,2015]
for i in years:
    print i, pop(i)

```

```python
# get the total motivations in just the cities
def city_totals(year):
    df = table_13[str(year)].copy()
    print cities(year)["pop"].sum()
    df = df[(df["Unnamed: 0"] == "Cities") & (df["Unnamed: 1"].isnull())]
    df["pop"] = cities(year)["pop"].sum()
    df["year"] = year
    return df
    
city_sums = pd.concat([city_totals(2010),
           city_totals(2011),
           city_totals(2012),
           city_totals(2013),
           city_totals(2014),
           city_totals(2015)])[[4,5,6,7,8,9,10,11,12,15,16]]
city_sums
```

```python
def adj(category):
    city_sums["adj_" + category] = city_sums[category].fillna(0) * 1000 * 1000 / city_sums["pop"]
    
adj("Disability")
adj("Race")
adj("Ethnicity")
adj("Race/\nEthnicity/\nAncestry")
adj("Religion")
adj('Sexual\norientation')

# city_sums["tot"] = city_sums["adj_Disability"] + city_sums["adj_Race"] + city_sums["adj_Ethnicity"]\
# + city_sums["adj_Race/\nEthnicity/\nAncestry"] + city_sums["adj_Religion"] + city_sums['adj_Sexual\norientation']

city_sums["tot_2"] = (city_sums["Disability"].fillna(0)\
+ city_sums["Race"].fillna(0) + city_sums["Ethnicity"].fillna(0)\
 + city_sums["Race/\nEthnicity/\nAncestry"].fillna(0)\
 + city_sums["Religion"].fillna(0) + city_sums['Sexual\norientation'].fillna(0)) * 1000 * 100\
/ city_sums["pop"] 

city_sums
```

```python

def totals(year):
    df = table_13[year].copy()
    df = df.sum().to_frame()
    df.columns = [year]
    return df
    
totals("2010")\
.join(totals("2011"),how="outer")\
.join(totals("2012"),how="outer")\
.join(totals("2013"),how="outer")\
.join(totals("2014"),how="outer")\
.join(totals("2015"),how="outer")
```
